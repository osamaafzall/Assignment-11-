```python
def sql_query(q):
    return pd.read_sql_query(q, conn)
```

# 1. Show All Employees With A Total Pay Between 50,000 and 75,000

```python
q = '''select employeename, Totalpay from salaries
    where totalpay between 50000 and 75000
    
'''
sql_query(q)
```

# 2. Show All Employees with A Basepay less than 50,000 Or A TotalPay Greater Than 100,000

```python
q = '''select employeename, Basepay from salaries
    where Basepay < 50000 or Totalpay > 100000'''

sql_query(q)
```

# 3. Show All Employees With A Totalpay Benefits Valus Between 125,000 and 150,000 And A job Title Containing word 'Director'

```python
q = '''select employeename, TotalpayBenefits, jobtitle from salaries
    where TotalpayBenefits between 125000 and 150000
    and 
    jobtitle LIKE '%Director%'

'''
sql_query(q)
```

# 4.Show All Employees Ordered By Their TotalPay Benefits in Descending Order

```python
q = 'select employeename, TotalpayBenefits from salaries order by TotalpayBenefits DESC'

sql_query(q)
```

# 5. How many employees in the salaries table have Overtimepay equal to 0

```python
q = '''select count(*) from salaries
    where Overtimepay = 0
'''
sql_query(q)
```
